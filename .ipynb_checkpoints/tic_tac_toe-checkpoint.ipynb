```python
def is_winner(board, player):
    for i in range(3):
        cnt=0
        cntC=0
        for j in range(3):
            if board[i][j]==player:  #check all columns for ith Row
                cnt+=1
            if board[j][i]==player:  #check all rows for ith column
                cntC+=1
        if cnt==3:
            return True
        elif cntC==3:
            return True
    if board[0][0]==player and board[1][1]==player and board[2][2]==player:
            return True
    elif board[0][2]==player and board[1][1]==player and board[2][0]==player:
            return True
    return False    
```

```python
def copy_board(board):
    b=[]
    for i in range(3):
        rr=[ board[i][j] for j in range(3) ]
        b.append(rr)
    return b
```

```python
# def is_free(board, r, c):
#     if board[r][c]==' ':
#         return True
#     return False
def is_free(board, pos):
    r,c=int(pos/3), pos%3
    if board[r][c]==' ':
        return True
    return False
```

```python
def count_free(board):
    nfree=0
    for i in range(9):
        if is_free(board, i):
            nfree+=1
    return nfree
```

```python
def ai_move(board):
    nfree=count_free(board)
    if nfree==0:
        return False
    while True:
        r=random.random()*8
        r=round(r)
        if is_free(board, r):
            r,c=int(r/3), r%3
            board[r][c]='O'
            break
    return True
```

```python
def finish_game(xw, xO):
    print('Game Finished---------')
    if xw:
        print("Winner=Human")
    elif xO:
        print('Winner=Random AI')
    else:
        print('Game Draw')
```

```python
class HumanAgent():
    def __init__(self, symbol='X'):
        print('human agent')
        self.symbol=symbol
    def next_move(self, board):
        nfree=board.count_free()
        if nfree==0:
            return False
        while True:
            inp=int( input('Enter cell No(1-9): ') )
            inp=inp-1 
            if board.is_free(inp):
                r,c=int(inp/3), inp%3
                board.board[r][c]=self.symbol
                break
            else:
                print('Please give valid input')
        return True
```

```python
class RandomAgent():
    def __init__(self, symbol='O'):
        print('random ai')
        self.symbol=symbol
    def next_move(self, board):
        nfree=board.count_free()
        if nfree==0:
            return False
        while True:
            r=random.random()*8
            r=round(r)
            if board.is_free(r):
                r,c=int(r/3), r%3
                board.board[r][c]=self.symbol
                break
        return True
```

```python
class TicTacToe():
    def __init__(self):
        print('tic-tac-toe game')
        self.verbose=True
    def start(self, tb, human, ai):
        tb.reset_board()
        human_turn=True   #First Human
        winner=None
        while True:
            if self.verbose:
                tb.print_board()
            win_or_draw, w=tb.is_game_over()
            if win_or_draw:
                winner=w
                break
            if human_turn:
                human.next_move(tb)
            else:
                ai.next_move(tb)
            human_turn=not human_turn
        if self.verbose:
            print('___GAME__COMPLETED___')
        return winner
    
```

```python
class MiniMaxAgent():
    def __init__(self, symbol='O'):
        print('MiniMax ai')
        self.symbol=symbol 
    def next_move(self, board):
        best,bp =self.minimax(board, self.symbol)
        print('best=',best, ' best_p=', bp)
        board.make_move(bp, 'O')
    def oponent(self, player):
        if player=='X':
            return 'O'
        else:
            return 'X'
        
    '''
    Have Some prob. 4,5,6
    '''
    def minimax(self, tb, player):
        go,w=tb.is_game_over()
        if go:
            score=self.get_score(w)
            return score, 0
        best=0
        best_p=None
        if player==self.symbol:
#             best=0 
            for m in tb.available_moves():
                tb.make_move(m, player)
#                 print('second_max:')
#                 tb.print_board()
                val, bp=self.minimax(tb, self.oponent(player) )
                tb.make_move(m, ' ')
                if val>=best:
                    best=val
                    best_p=m 
        elif player==self.oponent(self.symbol):
#             best=1000 
            for m in tb.available_moves():
                tb.make_move(m, player)
#                 print('second_min: ')
#                 tb.print_board()
                val,bp=self.minimax(tb, self.oponent(player) )
                tb.make_move(m, ' ')
                if val<=best:
                    best=val
                    best_p=m 
        return best, best_p
        
    def get_score(self, winner):
        if winner==self.oponent(self.symbol):
            return -1
        elif winner==self.symbol:
            return +1
        return 0
```
